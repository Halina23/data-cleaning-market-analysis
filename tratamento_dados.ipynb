```python
import os
import pandas as pd


# 1. CONFIGURAÇÃO

caminho_arquivo = "data/Resultados_Mercado.xlsx"
pasta_saida = "outputs"

os.makedirs(pasta_saida, exist_ok=True)


# 2. FUNÇÕES AUXILIARES

def limpar_colunas(df):
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace("\n", " ", regex=False)
    )
    df = df.loc[:, ~df.columns.str.contains("^Unnamed", na=False)]
    df = df.drop(columns=["Coluna1"], errors="ignore")
    return df

def arredondar_numericas(df):
    for col in df.columns[1:]:
        serie_convertida = pd.to_numeric(df[col], errors="coerce")
        if serie_convertida.notna().any():
            df[col] = serie_convertida.round(2)
    return df

def salvar_csv(df, nome_arquivo):
    df.to_csv(
        os.path.join(pasta_saida, nome_arquivo),
        index=False,
        sep=";",
        decimal=","
    )

def tratar_aba_simples(sheet_name, nome_csv, header=0):
    df = pd.read_excel(caminho_arquivo, sheet_name=sheet_name, header=header)
    df = limpar_colunas(df)
    df = arredondar_numericas(df)

    if len(df.columns) == 2:
        df.columns = [df.columns[0], "Porcentagem"]
        df["Porcentagem"] = pd.to_numeric(df["Porcentagem"], errors="coerce").round(2)

    salvar_csv(df, nome_csv)
    print(f"✅ {sheet_name} tratado e salvo em {nome_csv}")
    display(df.head(10))
    return df

def tratar_aba_matriz(sheet_name, nome_csv, id_var, header=2, var_name="Categoria", value_name="Porcentagem"):
    df = pd.read_excel(caminho_arquivo, sheet_name=sheet_name, header=header)
    df = limpar_colunas(df)

    df_longo = pd.melt(
        df,
        id_vars=[id_var],
        var_name=var_name,
        value_name=value_name
    )

    df_longo[value_name] = pd.to_numeric(df_longo[value_name], errors="coerce").round(2)

    salvar_csv(df_longo, nome_csv)
    print(f"✅ {sheet_name} tratado e salvo em {nome_csv}")
    display(df_longo.head(10))
    return df_longo


# 3. PERFIL E DINÂMICA DE USO

try:
    print("\n=== PERFIL ===")
    df_perfil = tratar_aba_simples("Perfil", "perfil_limpo.csv", header=0)

    print("\n=== USUÁRIO ===")
    df_usuario = tratar_aba_simples("Usuário", "usuario_limpo.csv", header=0)

    print("\n=== FAIXA ETÁRIA ===")
    df_faixa = tratar_aba_simples("Faixa Etária", "faixa_etaria_limpo.csv", header=0)

    print("\n=== IDENTIFICAÇÃO DE PERFIL ===")
    df_identificacao = tratar_aba_matriz(
        "Identificação de Perfil",
        "identificacao_perfil_limpo.csv",
        id_var="UF",
        header=2,
        var_name="Perfil",
        value_name="Porcentagem"
    )

    print("\n=== REGIÃO ===")
    df_regiao = tratar_aba_simples("Região", "regiao_limpo.csv", header=0)

    print("\n=== FREQUÊNCIA DE TIRO ===")
    df_freq = tratar_aba_simples("Frequência de Tiro", "frequencia_tiro_limpo.csv", header=0)

    print("\n=== NÍVEL TÉCNICO ===")
    df_nivel = tratar_aba_simples("Nível Técnico", "nivel_tecnico_limpo.csv", header=0)

    print("\n=== QNTDPROJÉTEIS ===")
    df_qtd = tratar_aba_simples("QntdProjéteis", "qntd_projeteis_limpo.csv", header=2)

    print("\n=== MODALIDADE ===")
    df_modalidade = tratar_aba_simples("Modalidade", "modalidade_limpo.csv", header=0)

    print("\n🚀 BLOCO 'PERFIL E DINÂMICA DE USO' PROCESSADO COM SUCESSO!")

except Exception as e:
    print(f"❌ Erro no processamento: {e}")
```

```python
import os
import pandas as pd


# 1. CONFIGURAÇÃO

caminho_arquivo = "data/Resultados_Mercado.xlsx"
pasta_saida = "outputs"

os.makedirs(pasta_saida, exist_ok=True)


# 2. FUNÇÕES AUXILIARES

def limpar_colunas(df):
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace("\n", " ", regex=False)
    )
    df = df.loc[:, ~df.columns.str.contains("^Unnamed", na=False)]
    df = df.drop(columns=["Coluna1"], errors="ignore")
    return df

def salvar_csv(df, nome_arquivo):
    df.to_csv(
        os.path.join(pasta_saida, nome_arquivo),
        index=False,
        sep=";",
        decimal=","
    )

def arredondar_numericas(df):
    for col in df.columns[1:]:
        serie = pd.to_numeric(df[col], errors="coerce")
        if serie.notna().any():
            df[col] = serie.round(2)
    return df


# 3. ESCOLHA PROJÉTEIS

try:
    # ---------------------------------
    # A) ESCOLHA PROJÉTEIS
    # ---------------------------------
    print("\n=== ESCOLHA PROJÉTEIS ===")
    df_escolha_proj = pd.read_excel(
        caminho_arquivo,
        sheet_name="Escolha Projéteis",
        header=2
    )

    df_escolha_proj = limpar_colunas(df_escolha_proj)
    df_escolha_proj.columns = ["Calibre", "Porcentagem"]
    df_escolha_proj["Porcentagem"] = pd.to_numeric(df_escolha_proj["Porcentagem"], errors="coerce").round(2)

    salvar_csv(df_escolha_proj, "escolha_projeteis_limpo.csv")
    print("✅ Escolha Projéteis salvo")
    display(df_escolha_proj)


    # B) NÍVEL TÉCNICO CALIBRE
  
    print("\n=== NÍVEL TÉCNICO CALIBRE ===")
    df_nivel_calibre = pd.read_excel(
        caminho_arquivo,
        sheet_name="Nível Técnico Calibre",
        header=2
    )

    df_nivel_calibre = limpar_colunas(df_nivel_calibre)
    df_nivel_calibre = arredondar_numericas(df_nivel_calibre)

    salvar_csv(df_nivel_calibre, "nivel_tecnico_calibre_limpo.csv")
    print("✅ Nível Técnico Calibre salvo")
    display(df_nivel_calibre)

  
    # C) CALIBRE REGIÃO
    
    print("\n=== CALIBRE REGIÃO ===")
    df_calibre_regiao = pd.read_excel(
        caminho_arquivo,
        sheet_name="Calibre Região",
        header=0
    )

    df_calibre_regiao = limpar_colunas(df_calibre_regiao)
    df_calibre_regiao = arredondar_numericas(df_calibre_regiao)

    salvar_csv(df_calibre_regiao, "calibre_regiao_limpo.csv")
    print("✅ Calibre Região salvo")
    display(df_calibre_regiao)

  
    # D) AVALIAÇÃO
  
    df_avaliacao = pd.read_excel(
        caminho_arquivo,
        sheet_name="Avaliação",
        header=2
    )

    df_avaliacao = limpar_colunas(df_avaliacao)

    # renomear colunas de nota
    df_avaliacao.columns = [
        "Fator",
        "Nota_1",
        "Nota_2",
        "Nota_3",
        "Nota_4",
        "Nota_5"
    ]

    df_avaliacao["Fator"] = df_avaliacao["Fator"].astype(str).str.strip()

    for col in ["Nota_1", "Nota_2", "Nota_3", "Nota_4", "Nota_5"]:
        df_avaliacao[col] = pd.to_numeric(df_avaliacao[col], errors="coerce").round(2)

    salvar_csv(df_avaliacao, "avaliacao_escolha_projetil_limpo.csv")
    print("✅ Avaliação salva")
    display(df_avaliacao)

    
    # E) IDENT.PERFIL
  
    print("\n=== IDENT.PERFIL ===")
    df_ident_perfil = pd.read_excel(
        caminho_arquivo,
        sheet_name="Ident.Perfil",
        header=0
    )

    df_ident_perfil = limpar_colunas(df_ident_perfil)
    df_ident_perfil.columns = [
        "Fator",
        "Usuario_Final",
        "Gestor_Clube",
        "Gestor_Loja_Armaria"
    ]

    df_ident_perfil["Fator"] = df_ident_perfil["Fator"].astype(str).str.strip()

    for col in ["Usuario_Final", "Gestor_Clube", "Gestor_Loja_Armaria"]:
        df_ident_perfil[col] = pd.to_numeric(df_ident_perfil[col], errors="coerce").round(2)

    salvar_csv(df_ident_perfil, "avaliacao_por_perfil_limpo.csv")
    print("✅ Ident.Perfil salvo")
    display(df_ident_perfil)

  
    # F) NÍVELTECNICO
    
    df_nivel_tecnico = pd.read_excel(
        caminho_arquivo,
        sheet_name="NívelTecnico",
        header=2
    )

    df_nivel_tecnico = limpar_colunas(df_nivel_tecnico)

    # a aba veio sem coluna de fator; reconstruir usando a ordem do Ident.Perfil
    fatores = df_ident_perfil["Fator"].tolist()

    # limitar ao número de fatores
    df_nivel_tecnico = df_nivel_tecnico.iloc[:len(fatores), :]

    # inserir coluna Fator
    df_nivel_tecnico.insert(0, "Fator", fatores)

    # renomear colunas
    df_nivel_tecnico.columns = [
        "Fator",
        "Competitivo_Alta_Performance",
        "Avancado",
        "Intermediario",
        "Iniciante"
    ]

    for col in ["Competitivo_Alta_Performance", "Avancado", "Intermediario", "Iniciante"]:
        df_nivel_tecnico[col] = pd.to_numeric(df_nivel_tecnico[col], errors="coerce").round(2)

    salvar_csv(df_nivel_tecnico, "avaliacao_por_nivel_tecnico_limpo.csv")
    print("✅ NívelTecnico salvo")
    display(df_nivel_tecnico)

    print("\n🚀 BLOCO 'ESCOLHA DO PROJÉTIL' PROCESSADO COM SUCESSO!")

except Exception as e:
    print(f"❌ Erro no processamento: {e}")
```

```python
import os
import pandas as pd


# CONFIGURAÇÃO

caminho_arquivo = "data/Resultados_Mercado.xlsx"
pasta_saida = "outputs"

os.makedirs(pasta_saida, exist_ok=True)


# FUNÇÕES AUXILIARES

def limpar_colunas(df):
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace("\n", " ", regex=False)
    )
    df = df.loc[:, ~df.columns.str.contains("^Unnamed", na=False)]
    df = df.drop(columns=["Coluna1"], errors="ignore")
    return df

def salvar_csv(df, nome):
    df.to_csv(
        os.path.join(pasta_saida, nome),
        index=False,
        sep=";",
        decimal=","
    )

def arredondar(df):
    for col in df.columns[1:]:
        serie = pd.to_numeric(df[col], errors="coerce")
        if serie.notna().any():
            df[col] = serie.round(2)
    return df

def tratar_simples(sheet, nome_csv, header=0):
    df = pd.read_excel(caminho_arquivo, sheet_name=sheet, header=header)
    df = limpar_colunas(df)
    df = arredondar(df)

    if len(df.columns) == 2:
        df.columns = [df.columns[0], "Porcentagem"]
        df["Porcentagem"] = pd.to_numeric(df["Porcentagem"], errors="coerce").round(2)

    salvar_csv(df, nome_csv)
    print(f"✅ {sheet} OK")
    display(df.head(10))
    return df

def tratar_matriz(sheet, nome_csv, header=2):
    df = pd.read_excel(caminho_arquivo, sheet_name=sheet, header=header)
    df = limpar_colunas(df)

    # pega automaticamente a primeira coluna
    base = df.columns[0]

    print(f"Colunas em {sheet}:", df.columns.tolist())

    df_longo = pd.melt(
        df,
        id_vars=[base],
        var_name="Categoria",
        value_name="Porcentagem"
    )

    df_longo["Porcentagem"] = pd.to_numeric(
        df_longo["Porcentagem"],
        errors="coerce"
    ).round(2)

    salvar_csv(df_longo, nome_csv)
    print(f"✅ {sheet} OK (matriz)")
    display(df_longo.head(10))
    return df_longo


# BLOCO: MARCAS + NOVO PROJÉTIL

try:

    print("\n=== MARCAS ===")
    tratar_simples("Marcas", "marcas_limpo.csv", header=0)

    print("\n=== REGIÕESMARCAS ===")
    tratar_matriz("RegiõesMarcas", "regioes_marcas_limpo.csv", header=2)

    print("\n=== MODALI.PROJÉTIL ===")
    tratar_simples("Modali.Projétil", "modali_projetil_limpo.csv", header=0)

    print("\n=== RECOM.PROJÉTEIS ===")
    tratar_matriz("Recom.Projéteis", "recom_projeteis_limpo.csv", header=2)

    print("\n=== TIPOS PROJÉTEIS ===")
    tratar_simples("Tipos Projéteis", "tipos_projeteis_limpo.csv", header=0)

    print("\n=== TESTARNOVO ===")
    tratar_simples("TestarNovo", "testar_novo_limpo.csv", header=0)

    print("\n=== MOTIVAÇÃONOVO ===")
    tratar_simples("MotivaçãoNovo", "motivacao_novo_limpo.csv", header=0)

    print("\n=== CONFIA.MARCAS ===")
    tratar_simples("Confia.Marcas", "confia_marcas_limpo.csv", header=0)

    print("\n=== CONFIA.PERFIL ===")
    df_confia_perfil = pd.read_excel(
        caminho_arquivo,
        sheet_name="Confia.Perfil",
        header=2
    )

    df_confia_perfil = limpar_colunas(df_confia_perfil)
    df_confia_perfil = arredondar(df_confia_perfil)

    salvar_csv(df_confia_perfil, "confia_perfil_limpo.csv")
    print("✅ Confia.Perfil OK")
    display(df_confia_perfil.head(10))

    print("\n=== ERROS MERCADO ===")
    tratar_simples("Erros Mercado", "erros_mercado_limpo.csv", header=0)

    print("\n=== REDE ===")
    tratar_simples("Rede", "rede_limpo.csv", header=0)

    print("\n🚀 BLOCO FINAL PROCESSADO COM SUCESSO!")

except Exception as e:
    print(f"❌ Erro: {e}")
```
